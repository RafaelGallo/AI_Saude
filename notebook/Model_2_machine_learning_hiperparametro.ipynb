# Pré-processamento 

O Label Encoder é uma técnica de pré-processamento de dados usada especificamente para lidar com a normalização de variáveis categóricas em machine learning. Variáveis categóricas são aquelas que representam categorias, como "maçã", "laranja" ou "banana", em oposição a variáveis numéricas que representam valores quantitativos. Modelos de machine learning geralmente esperam entradas numéricas, portanto, é necessário converter variáveis categóricas em valores numéricos. O LabelEncoder é uma das maneiras de realizar essa tarefa.

O LabelEncoder atribui um valor numérico único a cada categoria em uma variável categórica. Por exemplo, se você tiver uma coluna "Fruta" com valores "Maçã", "Laranja" e "Banana", o LabelEncoder atribuirá valores numéricos como 0, 1 e 2, respectivamente, a essas categorias.

# Treino e teste

Treino e teste são duas etapas fundamentais no desenvolvimento de modelos de aprendizado de máquina (machine learning) e em outras técnicas estatísticas. Essas etapas são usadas para avaliar a capacidade de generalização do modelo e medir seu desempenho em dados não vistos anteriormente.

A divisão entre treino e teste é necessária para evitar que o modelo memorize os dados de treinamento (overfitting) e não consiga generalizar bem para novos dados. Aqui está uma explicação detalhada sobre cada etapa:

A) Treinamento: Durante a etapa de treinamento, o modelo é alimentado com um conjunto de dados denominado conjunto de treinamento. Esse conjunto contém exemplos rotulados, onde as entradas (características) estão associadas a um rótulo conhecido (resultado esperado). O modelo usa esses exemplos para aprender os padrões e relações presentes nos dados e ajustar seus parâmetros internos. O objetivo do treinamento é encontrar os melhores parâmetros do modelo que minimizem a diferença entre as previsões do modelo e os rótulos conhecidos. Isso é feito por meio de algoritmos de otimização que ajustam gradualmente os parâmetros com base nas diferenças (erros) entre as previsões e os rótulos conhecidos.

B) Teste: Após o treinamento, é necessário avaliar o desempenho do modelo em dados não vistos anteriormente. Para isso, utiliza-se um conjunto de dados separado chamado conjunto de teste. Esse conjunto também contém exemplos rotulados, mas o modelo não teve acesso a eles durante o treinamento. O conjunto de teste é usado para fazer previsões com o modelo treinado e comparar essas previsões com os rótulos conhecidos. Isso permite medir a capacidade de generalização do modelo e avaliar seu desempenho em dados desconhecidos. Métricas como precisão, acurácia, recall e F1-score são comumente utilizadas para avaliar o desempenho do modelo durante a fase de teste.

É importante destacar que o conjunto de teste é usado apenas para avaliação final do modelo, não para ajustar seus parâmetros. A divisão entre conjunto de treinamento e conjunto de teste ajuda a estimar o desempenho esperado do modelo em situações reais, onde dados não rotulados estão disponíveis.

Além do treinamento e teste, é comum também utilizar uma terceira etapa chamada de validação. Nessa etapa, um conjunto de dados separado chamado conjunto de validação é utilizado para ajustar os hiperparâmetros do modelo, como taxa de aprendizado, número de camadas, tamanho do batch, etc. A validação auxilia na seleção dos melhores hiperparâmetros antes de finalizar o treinamento e avaliar o modelo no conjunto de teste.

A separação adequada entre treino, validação e teste é crucial para garantir uma avaliação justa e confiável do desempenho do modelo, além de evitar problemas de overfitting e subestimação de resultados.

# Modelo machine learning

Um modelo de machine learning é um algoritmo ou conjunto de algoritmos que é treinado para realizar tarefas específicas, como classificação, regressão, clustering ou previsão, com base em dados. Esses modelos são uma parte fundamental da inteligência artificial e da aprendizagem de máquina, que são campos da informática que se concentram em criar sistemas que podem aprender e tomar decisões com base em dados, em vez de serem explicitamente programados para realizar tarefas.

A ideia central por trás dos modelos de machine learning é que eles podem aprender a partir de exemplos e dados, em vez de depender de regras e programação explícita. ## Algoritmos bagging

# O que é Acúracia ?

A acurácia é uma métrica comumente usada para avaliar o desempenho de um modelo de aprendizado de máquina em problemas de classificação. Ela mede a proporção de instâncias classificadas corretamente em relação ao número total de instâncias no conjunto de dados. A fórmula básica para calcular a acurácia é:

Acurácia = (Número de instâncias classificadas corretamente) / (Número total de instâncias)

A acurácia é expressa como um valor entre 0 e 1, onde 1 representa uma acurácia perfeita, indicando que todas as instâncias foram classificadas corretamente, e 0 representa uma acurácia nula, indicando que nenhuma instância foi classificada corretamente. A acurácia é uma métrica simples e fácil de entender, sendo amplamente utilizada como uma medida inicial do desempenho de um modelo. No entanto, ela pode não ser adequada para todos os problemas, especialmente quando os conjuntos de dados estão desequilibrados, ou seja, quando uma classe é muito mais frequente do que a outra. Nesses casos, outras métricas, como precisão, recall ou F1-score, podem fornecer uma visão mais completa e precisa do desempenho do modelo. É importante ressaltar que a acurácia por si só não é suficiente para avaliar completamente a qualidade de um modelo. Outras métricas e considerações, como matriz de confusão, curvas ROC (Receiver Operating Characteristic) e validação cruzada, devem ser levadas em conta para obter uma compreensão mais abrangente do desempenho do modelo em diferentes aspectos.

# Classification report

O classification report é um relatório que fornece uma visão detalhada das métricas de desempenho de um modelo de classificação. Ele é especialmente útil para problemas de classificação multiclasse, onde o modelo tem a tarefa de classificar instâncias em mais de duas classes. O classification report geralmente inclui as seguintes métricas para cada classe do problema de classificação:

**Precision (precisão):** A precisão é a proporção de instâncias classificadas corretamente para uma determinada classe em relação ao número total de instâncias classificadas como essa classe. Ela mede a qualidade das previsões positivas para a classe em questão.


**Recall (revocação) ou Sensitivity (sensibilidade):** O recall é a proporção de instâncias classificadas corretamente para uma determinada classe em relação ao número total de instâncias reais dessa classe. Ele mede a capacidade do modelo de identificar corretamente as instâncias positivas para a classe em questão.


**F1-score:** O F1-score é uma medida que combina precisão e recall em uma única métrica. É a média harmônica dessas duas métricas e fornece uma medida geral do desempenho do modelo para uma determinada classe.


**Support (suporte):** O suporte é o número total de instâncias reais para uma determinada classe no conjunto de dados. Ele fornece informações sobre a distribuição das instâncias nas diferentes classes.


Além dessas métricas para cada classe, o classification report geralmente inclui a média dessas métricas, considerando todas as classes, o que dá uma visão geral do desempenho global do modelo. Essas médias podem ser calculadas de várias formas, como a média aritmética simples (macro-average), a média ponderada pelo suporte de cada classe (weighted average) ou a média ponderada pela frequência real de cada classe (micro-average). O classification report é uma ferramenta útil para avaliar o desempenho do modelo de classificação, pois oferece uma visão detalhada das métricas para cada classe individualmente e para o conjunto de dados como um todo. Ele permite identificar classes com desempenho inferior, verificar o equilíbrio das métricas em diferentes classes e tomar decisões informadas sobre ajustes no modelo, como ações de ajuste de limiar ou ponderação de classes desequilibradas.

```python
# Classification report
from sklearn.metrics import classification_report

classification = classification_report(model_random_forest_regressor_pred, y_test)
print("Modelo - Random forest")
print()
print(classification)
```

## Curva ROC (Receiver Operating Characteristic)

A curva ROC (Receiver Operating Characteristic) é uma ferramenta importante para avaliar o desempenho de modelos de classificação, como a regressão logística. A curva ROC ilustra a capacidade de um modelo de distinguir entre classes positivas e negativas em diferentes pontos de corte.

```python
# Importando biblioteca
from sklearn.metrics import roc_curve, roc_auc_score

# Calcule a curva ROC
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)

# Calcule a área sob a curva ROC (AUC-ROC)
roc_auc = roc_auc_score(y_test, y_pred_proba)

# Crie um gráfico da curva ROC
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'AUC = {roc_auc:.2f}')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Taxa de Falsos Positivos')
plt.ylabel('Taxa de Verdadeiros Positivos')
plt.title('Curva ROC')
plt.legend(loc='lower right')
plt.show()
```

# O que é confusion matrix ?

Confusion matrix (matriz de confusão) é uma tabela que descreve o desempenho de um modelo de classificação, fornecendo uma visão detalhada das previsões feitas pelo modelo em comparação com as classes reais do conjunto de dados. A matriz de confusão tem dimensões KxK, onde K representa o número de classes no problema de classificação. Cada célula da matriz de confusão fornece informações sobre a contagem ou frequência das instâncias classificadas em uma classe verdadeira (linha) em relação à classe prevista (coluna). Aqui estão os elementos principais de uma matriz de confusão:

**Verdadeiro Positivo (True Positive, TP):** O número de instâncias que foram corretamente classificadas como positivas para a classe verdadeira.

**Falso Positivo (False Positive, FP):** O número de instâncias que foram incorretamente classificadas como positivas para a classe verdadeira. Também é conhecido como erro do Tipo I.

**Verdadeiro Negativo (True Negative, TN):** O número de instâncias que foram corretamente classificadas como negativas para a classe verdadeira.

**Falso Negativo (False Negative, FN):** O número de instâncias que foram incorretamente classificadas como negativas para a classe verdadeira. Também é conhecido como erro do Tipo II.

A matriz de confusão permite visualizar facilmente os erros de classificação feitos pelo modelo. Ela fornece uma visão geral dos resultados do modelo, destacando as áreas em que o modelo está acertando e errando em relação a cada classe. Com base na matriz de confusão, várias métricas de desempenho podem ser calculadas, incluindo precisão (precision), recall (revocação), F1-score e acurácia. Essas métricas podem ajudar a avaliar o desempenho global do modelo e entender sua capacidade de fazer previsões corretas para diferentes classes.

A matriz de confusão é uma ferramenta importante para avaliar e diagnosticar o desempenho do modelo de classificação, especialmente em problemas multiclasse ou quando há desequilíbrio nas classes do conjunto de dados. Ela oferece informações mais detalhadas do que a acurácia sozinha, permitindo uma análise mais precisa das previsões do modelo e identificação de áreas para melhoria.

# Validação cruzada

A validação cruzada (cross-validation, em inglês) é uma técnica comum usada em estatísticas e aprendizado de máquina para avaliar o desempenho de um modelo preditivo ou classificador. Seu principal objetivo é estimar o quão bem o modelo será capaz de generalizar para dados não vistos, ou seja, para dados que não foram usados no treinamento.

A validação cruzada envolve a divisão do conjunto de dados em várias partes, comumente chamadas de "fold" ou dobras, e a realização de múltiplos treinamentos e testes do modelo em diferentes combinações dessas dobras. O procedimento típico envolve os seguintes passos.

**Divisão dos dados:** O conjunto de dados é dividido em k dobras, onde k é um número inteiro especificado (geralmente 5 ou 10, mas pode variar). Cada dobra contém uma parte dos dados.

**Treinamento e teste:** O modelo é treinado k vezes, cada vez usando k-1 dobras para treinamento e a dobra restante para teste. Isso gera k conjuntos de métricas de desempenho (como acurácia, erro médio quadrático, etc.).

**Avaliação do desempenho:** O desempenho do modelo é avaliado calculando métricas de desempenho (por exemplo, média ou mediana das métricas de desempenho) com base nos resultados dos testes em cada dobra.

A validação cruzada é útil porque fornece uma estimativa mais robusta e confiável do desempenho do modelo em comparação com uma única divisão de treinamento e teste. Isso ajuda a evitar a superestimação ou subestimação do desempenho do modelo devido à aleatoriedade na divisão dos dados. Além disso, a técnica permite que todo o conjunto de dados seja usado tanto para treinamento quanto para teste, o que é particularmente importante quando o conjunto de dados é pequeno.

Um dos tipos mais comuns de validação cruzada é a validação cruzada k-fold (k-fold cross-validation), que foi descrita acima. Outras variações incluem a validação cruzada leave-one-out (LOOCV), em que cada ponto de dados é usado como teste uma vez, e a validação cruzada estratificada (stratified cross-validation), que preserva a distribuição das classes em cada dobra. A escolha do tipo de validação cruzada depende do problema e dos dados específicos em questão.

```python
%%time

# Importando biblioteca
from sklearn.model_selection import cross_validate

# Criando validação cruzada
results = cross_validate(model_random_forest,
                         X_train, 
                         y_train, 
                         cv=5,
                         scoring=('accuracy'),
                         return_train_score=True)

print(f"Mean train score {np.mean(results['train_score']):.2f}")
print(f"Mean test score {np.mean(results['test_score']):.2f}")
```

# Modelo machine learning - Hiperparâmetros com GridSearchCV

O GridSearchCV é uma técnica de busca em grade (grid search) usada para otimizar hiperparâmetros em modelos de machine learning. Ele é uma funcionalidade da biblioteca scikit-learn, que é amplamente utilizada para tarefas de aprendizado de máquina em Python.

A busca em grade é uma abordagem sistemática para encontrar a melhor combinação de hiperparâmetros de um modelo de machine learning. Hiperparâmetros são configurações do modelo que não são aprendidas durante o treinamento, mas precisam ser definidas antes de treinar o modelo, como a taxa de aprendizado, o número de vizinhos em um modelo K-Nearest Neighbors, ou o tipo de kernel em uma Máquina de Vetores de Suporte (SVM).

O GridSearchCV executa a validação cruzada em todas as combinações possíveis de hiperparâmetros que você especifica, ajudando a encontrar a configuração que produz o melhor desempenho do modelo.

Aqui estão os principais componentes do GridSearchCV:

Estimador do Modelo: Você fornece o modelo de machine learning que deseja otimizar (por exemplo, um classificador de Regressão Logística, uma Máquina de Vetores de Suporte, etc.).

Espaço de Hiperparâmetros: Você define um conjunto de valores ou opções para os hiperparâmetros que deseja otimizar. Por exemplo, para a Regressão Logística, pode ser a escolha entre diferentes valores de regularização (C) e tipos de penalização (l1 ou l2).

Método de Validação Cruzada: Você escolhe um método de validação cruzada, como k-fold cross-validation, que o GridSearchCV usará para avaliar o desempenho de cada combinação de hiperparâmetros.

Função de Pontuação: Você especifica uma métrica de avaliação que o GridSearchCV usará para classificar as diferentes configurações de hiperparâmetros. Por exemplo, pode ser a acurácia, F1-score, AUC-ROC, etc.

Número de Jobs em Paralelo: Você pode configurar o número de tarefas em paralelo para acelerar o processo de busca em grade, o que é útil em conjuntos de dados grandes.

O GridSearchCV então executa a validação cruzada em todas as combinações de hiperparâmetros e retorna a melhor combinação com base na métrica de avaliação escolhida.

```python
%%time

# Importando biblioteca
from joblib import parallel_backend
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier

# Defina o espaço de hiperparâmetros
param_grid = {
    'n_estimators': [100, 300, 500],  # Número de árvores na floresta
    'max_depth': [10, 20, 30],       # Profundidade máxima de cada árvore
    'min_samples_split': [2, 5, 10], # Número mínimo de amostras necessárias para dividir um nó
    'min_samples_leaf': [1, 2, 4],   # Número mínimo de amostras necessárias em uma folha
}

# Função personalizada para treinamento e GridSearchCV com serialização joblib
def train_model(params):
    with parallel_backend("loky"):
        rf_model = RandomForestClassifier(random_state=42, **params)
        return rf_model.fit(X_train, y_train)

# Crie um modelo Random Forest com GridSearchCV
grid_search = GridSearchCV(estimator=RandomForestClassifier(random_state=42),
                           param_grid=param_grid,
                           cv=5, scoring='accuracy', n_jobs=-1)

grid_search.fit(X_train, y_train)

# Obtenha a melhor combinação de hiperparâmetros
best_params = grid_search.best_params_

# Treine o modelo com os melhores hiperparâmetros usando a função personalizada
best_rf_model = train_model(best_params)

# Visualizando modelo
grid_search
```

```python
# Avalie o modelo
accuracy_GridSearchCV = best_rf_model.score(X_test, y_test)
print(f"Acurácia do modelo otimizado (GridSearchCV com Progresso): {accuracy_GridSearchCV:.2f}")
```

```python
# Previsão modelo com função predict de previsão 

best_rf_model_pred = best_rf_model.predict(X_test)
best_rf_model_pred
```

```python
# Faça previsões de probabilidade
y_pred_proba = best_rf_model.predict_proba(X_test)[:, 1]
y_pred_proba
```

```python
# Classification report
from sklearn.metrics import classification_report

classification = classification_report(best_rf_model_pred, y_test)
print("Modelo GridSearchCV - Random Forest Classifier")
print()
print(classification)
```

```python
# Curva ROC

# Importando biblioteca
from sklearn.metrics import roc_curve, roc_auc_score

# Calcule a curva ROC
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)

# Calcule a área sob a curva ROC (AUC-ROC)
roc_auc = roc_auc_score(y_test, y_pred_proba)

# Crie um gráfico da curva ROC
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'AUC = {roc_auc:.2f}')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('Taxa de Falsos Positivos')
plt.ylabel('Taxa de Verdadeiros Positivos')
plt.title('Curva ROC')
plt.legend(loc='lower right')
plt.show()
```

```python
# Importando biblioteca
from sklearn.metrics import confusion_matrix

# Confusion matrix 
matrix_confusion_1 = confusion_matrix(y_test, best_rf_model_pred)

print('Confusion matrix - Model GridSearchCV - Random Forest \n\n', matrix_confusion_1)
print('\nTrue Positives(TP) = ', matrix_confusion_1[0,0])
print('\nTrue Negatives(TN) = ', matrix_confusion_1[1,1])
print('\nFalse Positives(FP) = ', matrix_confusion_1[0,1])
print('\nFalse Negatives(FN) = ', matrix_confusion_1[1,0])

# Plot matrix

plt.figure(figsize=(10, 5))
ax = plt.subplot()
sns.heatmap(matrix_confusion_1, annot=True, ax = ax, fmt = ".1f", cmap="Paired"); 
ax.set_title('Confusion Matrix - Model GridSearchCV - Random Forest'); 
ax.xaxis.set_ticklabels(["Benigno", "Maligno"]); ax.yaxis.set_ticklabels(["Benigno", "Maligno"]);
```

# Modelo 2 Randomized Search CV

O RandomizedSearchCV é uma técnica de busca de hiperparâmetros que difere do GridSearchCV em sua abordagem para encontrar as melhores configurações de hiperparâmetros para um modelo de machine learning. Em vez de testar todas as combinações possíveis de hiperparâmetros, o RandomizedSearchCV amostra aleatoriamente dentro de um espaço de hiperparâmetros especificado, o que o torna mais eficiente para otimizar modelos com muitos hiperparâmetros.

Aqui estão os principais componentes e características do RandomizedSearchCV:

Espaço de Hiperparâmetros: Você especifica um espaço de hiperparâmetros que deseja otimizar. Em vez de listar todas as combinações possíveis, você define uma distribuição para cada hiperparâmetro. Por exemplo, você pode especificar uma distribuição uniforme ou discreta para um hiperparâmetro.

Número de Iterações: Você define o número de iterações ou amostras a serem retiradas do espaço de hiperparâmetros. O RandomizedSearchCV escolherá aleatoriamente essa quantidade de configurações para avaliar.

Estimador do Modelo: Você fornece o modelo de machine learning que deseja otimizar (por exemplo, um classificador Random Forest, Regressão Logística, etc.).

Método de Validação Cruzada: Você escolhe um método de validação cruzada, como k-fold cross-validation, que o RandomizedSearchCV usará para avaliar o desempenho de cada configuração.

Função de Pontuação: Você especifica a métrica de avaliação que o RandomizedSearchCV usará para classificar as diferentes configurações de hiperparâmetros.

Amostragem Aleatória: O RandomizedSearchCV seleciona configurações de hiperparâmetros de maneira aleatória dentro do espaço definido. Isso o torna mais eficiente, especialmente quando o espaço de busca é grande.

Eficiência: Em comparação com o GridSearchCV, o RandomizedSearchCV é mais eficiente para otimizar modelos com muitos hiperparâmetros, pois não testa todas as combinações possíveis.

Resultados Variáveis: Devido à natureza aleatória da amostragem, os resultados podem variar de uma execução para outra. No entanto, ele geralmente convergirá para uma configuração que fornece bom desempenho médio.

A principal vantagem do RandomizedSearchCV é que ele economiza tempo computacional ao amostrar apenas um número limitado de configurações de hiperparâmetros. No entanto, é importante notar que, devido à amostragem aleatória, pode não garantir que a melhor configuração seja encontrada, mas é mais eficiente do que o GridSearchCV em muitos casos.

Em resumo, o RandomizedSearchCV é uma ferramenta útil para otimização de hiperparâmetros quando você deseja economizar tempo e recursos computacionais, ao mesmo tempo em que busca configurações de hiperparâmetros eficazes para o seu modelo de machine learning.

```python
%%time

# Importando biblioteca
import scipy
from sklearn.model_selection import RandomizedSearchCV

# Defina o espaço de hiperparâmetros para RandomizedSearchCV
param_dist = {
    'n_estimators': [100, 300, 500],
    'max_depth': scipy.stats.randint(10, 30),
    'min_samples_split': scipy.stats.randint(2, 10),
    'min_samples_leaf': scipy.stats.randint(1, 4),
}

# Crie um modelo Random Forest
rf_model = RandomForestClassifier(random_state=42)

# Crie um objeto RandomizedSearchCV
random_search = RandomizedSearchCV(rf_model, 
                                   param_distributions=param_dist, 
                                   n_iter=10, 
                                   cv=5, 
                                   scoring='accuracy', 
                                   random_state=42)

# Realize a pesquisa aleatória para encontrar a melhor combinação de hiperparâmetros
random_search.fit(X_train, y_train)

# Encontre a melhor combinação de hiperparâmetros
best_params = random_search.best_params_

# Treine o modelo com os melhores hiperparâmetros
best_rf_model_2 = RandomForestClassifier(**best_params, random_state=42)
best_rf_model_2.fit(X_train, y_train)

# Visualizando modelo
best_rf_model_2
```

```python
# Avalie o modelo
accuracy_RandomizedSearchCV = best_rf_model_2.score(X_test, y_test)
print(f"Acurácia do modelo otimizado (RandomizedSearchCV): {accuracy_RandomizedSearchCV:.2f}")
```
